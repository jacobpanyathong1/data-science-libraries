```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

%config InlineBackend.figure_format='retina'
```

```python
x = np.arange(-4, 5, 0.1)
y = x**2 -x + 2

plt.plot(x, y, linewidth=2.0)

plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = (2**x)
plt.plot(x, y)
```

```python
def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/(x+1)
x=np.linspace(-10,10,101)
y=f(x)
plt.plot(x,y)
plt.show()
```

```python
# y=√x
x = np.arange(0, 5, 0.1)
y = np.emath.sqrt(x)

fig, ax = plt.subplots()
ax.plot(y)
ax.set_title('$y=√x$')

x = np.arange(-5, 5, 0.1)
y = (x**3)
fig, ax1 = plt.subplots()
ax1.plot(x, y)
ax1.set_title('$y=x^3$')

x = np.arange(0 , 5, 0.1)
y = (2**x)
fig, ax2 = plt.subplots()
ax2.plot(x, y)
ax2.set_title('$y=2^x$')

def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/(x+1)
x=np.linspace(-10,10,101)
y=f(x)
fig, ax3 = plt.subplots()
ax3.plot(x, y)
ax3.set_title('$y=1/(x+1)$')
```

```python
###
x = range(0, 10)
plt.figure(figsize=(14,6))
y = np.emath.sqrt(x)
plt.plot(x, y, c='dodgerblue', label='$y=sqrt(x)$')

###
x = np.arange(-5, 5, 0.1)
y = (x**3)
plt.plot(x, y,  c='firebrick', label='$y=x^3$')
###
y = (2**x)
plt.plot(x, y, c='seagreen', label='$y=2^x$')
###
def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/(x+1)
x=np.linspace(-10,10,101)
y=f(x)
plt.plot(x, y, c='darkorange', label='$y=1/(x+1)$')

plt.legend(loc='upper right')
plt.title('combined charts')
plt.xlabel('$x$')

plt.show()
```
