```python
import matplotlib.pyplot as plt
import numpy as np


# make data
x = np.linspace(0, 10, 100)
y = 1/(x+1)

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(-10, 10), xticks=np.arange(1, 8),
       ylim=(-10, 10), yticks=np.arange(1, 8))

plt.show()
```
