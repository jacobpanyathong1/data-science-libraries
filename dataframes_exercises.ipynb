```python
from pydataset import data
import pandas as p
import numpy as np
```

```python
# data('mpg', show_doc=True) # view the documentation for the dataset
mpg = data('mpg') # load the dataset and store it in a variable
```

```python
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']
math_grades = np.random.randint(low=60, high=100, size=len(students))

english_grades = np.random.randint(low=60, high=100, size=len(students))

reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = p.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
df
```

```python
df.assign(passing_english=df.english >= 75)
```

```python
df.assign(passing_english=df.english >= 75).sort_values(by='passing_english', ascending=False)
```

```python
df.assign(passing_english=df.english >= 75).sort_values(by=['passing_english', 'name'], ascending=[1, 1])
```

```python
df.assign(passing_english=df.english >= 75).sort_values(by=['passing_english', 'english'], ascending=[1, 0])
```

```python
df.assign(overall_gpa=(df['english']+df['math']+df['reading'])/3)
```

```python
mpg.assign(better_cty_then_hwy = mpg.cty > mpg.hwy).sort_values(by='better_cty_then_hwy', ascending=False)
```

```python
mpg.assign(mileage_difference = mpg.hwy - mpg.cty)
```

```python
mpg.assign(mileage_difference = mpg.hwy - mpg.cty).sort_values(by='mileage_difference', ascending=False).head(2)
```

```python
mpg.assign(average_mileage = (mpg.hwy + mpg.cty)/ 2).sort_values(by='average_mileage', ascending=False).head(1) # highest average compact car
```

```python
mpg.loc[mpg['class'] == 'compact'].nsmallest(1, 'hwy', keep='all')  #The best compact car is the VW Jetta with 23 hwy mileage
```

```python
mpg.loc[mpg['class'] =='compact'].nlargest(1, 'hwy', keep='all')  #The best compact car is the VW Jetta with 44 hwy mileage
```
